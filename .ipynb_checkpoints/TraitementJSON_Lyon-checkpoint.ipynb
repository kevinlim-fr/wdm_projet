```python
df_transport3 = pd.DataFrame()

for i in range(len(df_transport['geometry'])):
    dic3 = df_transport['geometry'][i]
    df_transport3 = df_transport3.append(dic3,ignore_index = True)

df_transport3
```

```python
coord= pd.DataFrame()
for i in range(len(df_transport3)):
    lat = df_transport3["coordinates"][i][0]
    lng = df_transport3["coordinates"][i][1]

    list_to_append = [lat,lng]
    line_labels = ['lat'] + ['lng'] 
    dic4 = {line_labels[i]: list_to_append[i] for i in range(len(list_to_append))}   
    coord = coord.append(dic4,ignore_index=True)
    
coord
```
