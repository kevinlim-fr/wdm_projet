```python
import requests

url_lyon_bike = 'https://download.data.grandlyon.com/wfs/rdata?SERVICE=WFS&VERSION=1.1.0&outputformat=GEOJSON&request=GetFeature&typename=jcd_jcdecaux.jcdvelov&SRSNAME=urn:ogc:def:crs:EPSG::4171'
data_lyon = requests.get(url_lyon_bike).json()



url_lyon_transport = 'https://download.data.grandlyon.com/wfs/rdata?SERVICE=WFS&VERSION=2.0.0&request=GetFeature&typename=tcl_sytral.tclarret&outputFormat=application%2Fjson%3B%20subtype%3Dgeojson&SRSNAME=EPSG%3A4171&fbclid=IwAR1F7a9zbsGvW49Bwj4cZKEOha8zEUrjZ4z8zk40R-7D4L_GnwvBY5D0ojM'
data_lyon_transport = requests.get(url_lyon_transport).json()


data_lyon_transport

url_hospital='https://raw.githubusercontent.com/kevinlim-fr/wdm_projet/main/originalJSON/les_etablissements_hospitaliers_franciliens.json'
data_hospital=requests.get(url_hospital).json()

```

```python

for element in data_lyon['features']:
    del element['properties']['number']
    del element['properties']['address']
    del element['properties']['address2']
    del element['properties']['commune']
    del element['properties']['nmarrond']
    del element['properties']['pole']
    del element['properties']['bonus']
    del element['properties']['status']
    del element['properties']['availabilitycode']
    del element['properties']['availability']
    del element['properties']['banking']
    del element['properties']['gid']
    del element['properties']['last_update']
    del element['properties']['last_update_fme']
    del element['properties']['code_insee']
    del element['properties']['langue']
    del element['properties']['nature']
    del element['properties']['titre']
    del element['properties']['description']
    del element['properties']['startdate']
    del element['properties']['enddate']
    del element['properties']['etat']
    del element['properties']['bike_stands']
    del element['properties']['available_bike_stands']
    del element['properties']['available_bikes']
    del element['geometry']
    element['properties']['city'] = 'Lyon'
    element['properties']['lat']=float(element['properties']['lat'])
    element['properties']['lng']=float(element['properties']['lng'])
    element['properties']['@type']='bike_station'
    
for element in data_lyon_transport['features']:
    del element['properties']['id']
    del element['properties']['desserte']
    del element['properties']['pmr']
    del element['properties']['ascenseur']
    del element['properties']['escalator']
    del element['properties']['gid']
    del element['properties']['last_update']
    del element['properties']['last_update_fme']
    element['properties']['name']= element['properties']['nom']
    del element['properties']['nom']
    element['properties']['city'] = 'Lyon'
    element['properties']['lat']=element['geometry']['coordinates'][1]
    element['properties']['lng']=element['geometry']['coordinates'][0]
    element['properties']['@type']='transport_station'
    del element['geometry']
    
      
```
